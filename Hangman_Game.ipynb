# **HANGMAN-GAME**

```python
import random

# Word list for the game
word_list = [
    "atlantic",
    "python",
    "programming",
    "challenge",
    "developer",
    "algorithm"
]

# ASCII art for the hangman
HANGMAN_ART = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---
    """,
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---
    """,
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---
    """,
    """
     ------
     |    |
     |    O
     |   /|\
     |
     |
    ---
    """,
    """
     ------
     |    |
     |    O
     |   /|\
     |   /
     |
    ---
    """,
    """
     ------
     |    |
     |    O
     |   /|\
     |   / \
     |
    ---
    """
]

def choose_word():
    """Select a random word from the word list."""
    return random.choice(word_list)

def display_word(word, guessed_letters):
    """Display the word with guessed letters revealed and others as underscores."""
    return " ".join([letter if letter in guessed_letters else "_" for letter in word])

def hangman():
    """Main function to play the Hangman game."""
    word = choose_word()
    guessed_letters = set()
    attempts = 6  # Number of lives
    print("Welcome to Hangman!ğŸ¦¥")
    print("GuessğŸ§  the word and safeğŸ¦¸ğŸ» the manğŸ•´ğŸ»,before you run out of attemptsğŸ˜¿.")
    print(display_word(word, guessed_letters))

    while attempts > 0:
        print(HANGMAN_ART[6 - attempts])  # Display the hangman art
        guess = input("\nEnter a letter: ").lower()

        # Validate input
        if len(guess) != 1 or not guess.isalpha():
            print("âœ–ï¸ Please enter a single valid letterğŸ‘½.")
            continue

        # Check if the letter has already been guessed
        if guess in guessed_letters:
            print("âŒ You've already guessed that letter. Try again!â˜¹ï¸.")
            continue

        guessed_letters.add(guess)

        # Check if the guessed letter is in the word
        if guess in word:
            print(f"ğŸŒŸ Good job! '{guess}' is in the word.")
        else:
            attempts -= 1
            print(f"ğŸ‘ Sorry, '{guess}' is not in the wordğŸ˜•. You have {attempts} attempts leftğŸ¥¶.")

        # Display the current state of the word
        current_display = display_word(word, guessed_letters)
        print(current_display)

        # Check if the player has guessed the entire word
        if "_" not in current_display:
            print("ğŸ‰ğŸ¥³Congratulations! You guessed the word!ğŸ‰and have saved the man!ğŸ¦¸")
            return

    # If the player runs out of attempts
    print(HANGMAN_ART[6])  # Display the final hangman art
    print(f"ğŸ˜­Game over! The word was '{word}'. Better luck next time!â˜ ï¸")

# Run the game
if __name__ == "__main__":
    hangman()
```
